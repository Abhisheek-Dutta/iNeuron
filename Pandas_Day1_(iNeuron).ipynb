```python
pd.read_excel('airline1.xls') #-> Read excel data set
```

```python
pd.read_excel('airline1.xls', sheet_name = "Sheet2") #-> Read specific Sheet2 of airline1 dataset
```

```python
# In Sheet3 there is no header or column name 

pd.read_excel('airline1.xls', sheet_name = "Sheet3")
```

```python
# In Sheet3 there is no header or column name 

pd.read_excel('airline1.xls', sheet_name = "Sheet3", header = None) 

#-> Using header = None function to assign indexes as column names
```

```python
pd.read_excel('airline1.xls', sheet_name = "Sheet3", header =  None , names = ['a', ' b', 'c', 'd', 'e', 'f'])

#-> Using header = None, names = ['a', ' b', 'c', 'd', 'e', 'f']) function to renaming the column names

#While renaming the columns. The names have to be equavalent to the columns
```

```python
pd1= pd.read_excel('airline1.xls', sheet_name = "Sheet3", header =  None , names = ['a', ' b', 'c', 'd', 'e', 'f'])
```

```python
pd1
```

```python
pd1.to_csv('Abhi.csv') #-> to save a file in csv

#pd1.to -> Hold the curser after to, for some time then you can select the file type to save your file
```

```python
pd1.to_csv('Abhi.csv', index = False) #-> to save a file and remove indexes

#pd1.to -> Hold the curser after to, for some time then you can select the file type to save your file
#index = False -> To remove indexes
```

```python
pd1.to_excel('Abhi.xlsx') #-> to save a file in excel
```

```python
pd1.to_csv('Abhi.csv', index = False, header = False) #-> to save a file and remove indexes also headers

#pd1.to -> Hold the curser after to, for some time then you can select the file type to save your file
#index = False -> To remove indexes
#header = False > To remove headers
```
