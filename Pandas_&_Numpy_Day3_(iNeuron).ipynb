```python
d2 = {"key1" : [1,2,3,4,5,"Abhisheek"],
          "key2" : [7,8,9,"Dutta",11,12], 
          "key3" : [13,14,15,16,17,"Data-Science"],
          "key4" : [19,20,"Abhi",22,23,24]
}
```

```python
df2 = pd.DataFrame(d2)
```

```python
df2
```

```python
df1
```

```python
pd.merge(df1, df2)

#-> pd.merge() - By default it does inner merge operation

#innner merge - Use intersection of keys from both frames, smiliar to a SQL inner join; preserve the 
#order of the left keys.

#inner merge gives outcome which common to two data sets and if not, it does not show that outcome 

#Here we will consider df1 as Left and df2 as Right

#On the beow output, the inner merge operation is reflecting only those values in a row which are 
#common in df1 and df2 datasets both.
```

```python
pd.merge(df1, df2, how = "inner", on = "key4")

# on function ->which column/key are supposed to be checked
```

```python
pd.merge(df1, df2, how = "left", on = "key4")
```

```python
pd.merge(df1, df2, how = "right", on = "key4")
```

```python
pd.merge(df1, df2, how = "outer", on = "key4")

#outer merge operation -> Use of union of keys from both frames, similar to a SQL full outer join; 
#sort keys lexicographically

#It brings every combinations from Right and Left
```

```python
pd.merge(df1, df2, how = "outer", on = ["key4", "key1"])
```
